### Forward propagation with Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
from builtins import range
```

```python
Nclass = 500

X1 = np.random.randn(Nclass, 2) + np.array([0, -2])
X2 = np.random.randn(Nclass, 2) + np.array([2, 2])
X3 = np.random.randn(Nclass, 2) + np.array([-2, 2])
X4 = np.random.randn(Nclass, 2) + np.array([-4, -2])
X = np.vstack([X1, X2, X3, X4])

Y = np.array([0]*Nclass + [1]*Nclass + [2]*Nclass + [3]*Nclass)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, s=50, alpha=0.5)
plt.show()
```

```python
# randomly initialize weights
D = 2 # dimensionality of input
M = 3 # hidden layer size
K = 4 # number of classes

W1 = np.random.randn(D, M)
b1 = np.random.randn(M)
W2 = np.random.randn(M, K)
b2 = np.random.randn(K)
```

```python
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
```

```python
def forward(X, W1, b1, W2, b2):
    Z = sigmoid(X.dot(W1) + b1) # sigmoid
    # Z = np.tanh(X.dot(W1) + b1) # tanh
    # Z = np.maximum(X.dot(W1) + b1, 0) # relu
    A = Z.dot(W2) + b2
    expA = np.exp(A)
    Y = expA / expA.sum(axis=1, keepdims=True)
    return Y
```

```python
def classification_rate(Y, P):
    n_correct = 0
    n_total = 0
    for i in range(len(Y)):
        n_total += 1
        if Y[i] == P[i]:
            n_correct += 1
    return float(n_correct) / n_total
```

```python
P_Y_given_X = forward(X, W1, b1, W2, b2)
P = np.argmax(P_Y_given_X, axis=1)
```

```python
assert(len(P) == len(Y))
```

```python
print("Classification rate for randomly choosen weights: ", classification_rate(Y, P))
```
